Importing the libraries.
Loading the dataset into dataframe.
Removing unnecessary rows from datasets. These are such rows which will lead us to wrong output.
Ensuring only necessary data is analyzed.
Identifying necessary columns .
Grouping by state and computing sum of crimes.
Creating a new column, "Total_Crimes", which represents the total number of crimes against women per state.
Finding the 5 states with lowest crime rtaes against womens.
PLotting the graph.
 Plots a horizontal bar chart (plt.barh()) with:
States on the Y-axis
Total crimes on the X-axis
ðŸ”¹ The safest state (least crimes) appears at the top using invert_yaxis().
ðŸ”¹ A gridline is added for clarity.




```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\Sai\Desktop\Q9\CAW merged data.xlsx"

df = pd.read_excel(file_path)

df = df[~df["DISTRICT"].str.contains("TOTAL", case=False, na=False)]
)
crime_columns = [col for col in df.columns if df[col].dtype in ['int64', 'float64']]


statewise_crime = df.groupby("STATE/UT")[crime_columns].sum().reset_index()


statewise_crime["Total_Crimes"] = statewise_crime[crime_columns].sum(axis=1)


safest_states = statewise_crime.nsmallest(5, "Total_Crimes")


plt.figure(figsize=(10, 5))
plt.barh(safest_states["STATE/UT"], safest_states["Total_Crimes"], color='green')
plt.xlabel("Total Crimes Against Women")
plt.ylabel("STATE/UT")
plt.title("Top 5 Safest States for Women (Lowest Crimes)")
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
