```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\Sai\Desktop\Q9\CAW merged data.xlsx"  
df_caw = pd.read_excel(file_path)


df_caw_cleaned = df_caw[~df_caw["DISTRICT"].isin(["TOTAL", "ZZ TOTAL"])]


df_caw_cleaned["Total_Crimes"] = df_caw_cleaned.drop(columns=["STATE/UT", "DISTRICT"]).sum(axis=1, numeric_only=True)

top_cities = df_caw_cleaned.groupby("DISTRICT")["Total_Crimes"].sum().nlargest(5).reset_index()


plt.figure(figsize=(10, 6))
plt.bar(top_cities["DISTRICT"], top_cities["Total_Crimes"], color='lightcoral')
plt.xlabel("District")
plt.ylabel("Total Crimes Against Women")
plt.title("Top 5 Cities with Highest Crimes Against Women")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

df_caw_cleaned.to_excel("CAW_cleaned_data.xlsx", index=False)
```
