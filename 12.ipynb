```python
import pandas as pd
import matplotlib.pyplot as plt

file_path_ipc = r"C:\Users\Sai\OneDrive\Desktop\Raw data\08_01_Juvenile_apprehended_state_IPC.csv"
file_path_sll = r"C:\Users\Sai\OneDrive\Desktop\Raw data\08_02_Juvenile_apprehended_state_SLL.csv"

df_ipc = pd.read_csv(file_path_ipc)
df_sll = pd.read_csv(file_path_sll)


columns_to_remove = ["Total for boys all Age Groups", "Total for girls all Age Groups", "Grand total"]
df_ipc = df_ipc.drop(columns=columns_to_remove)
df_sll = df_sll.drop(columns=columns_to_remove)


df_ipc_grouped = df_ipc.groupby(["STATE/UT", "CRIME"]).sum(numeric_only=True).reset_index()
df_sll_grouped = df_sll.groupby(["STATE/UT", "CRIME"]).sum(numeric_only=True).reset_index()


df_combined = pd.concat([df_ipc_grouped, df_sll_grouped])


top_states = df_combined.groupby("STATE/UT").sum(numeric_only=True).sum(axis=1).sort_values(ascending=False).head(10)


plt.figure(figsize=(10, 5))
top_states.plot(kind="bar", color="pink", edgecolor="black")
plt.xlabel("State")
plt.ylabel("Number of Cases")
plt.title("Top 10 States with Highest Juvenile Crimes")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
