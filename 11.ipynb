Imorting libraries.
Loading datasets.
The delimiter="\t" is used for economic_df because the file may be tab-separated instead of comma-separated.
Grouping data by year to anaylse trrnds over time.
Summing the juvlines arrested per year.
PLotting graph.
X axis: Year
Yaxis:Number of juvelines arrested.
Creating line plot.


```python
import pandas as pd
import matplotlib.pyplot as plt


education_df = pd.read_csv(r"C:\Users\Sai\OneDrive\Desktop\Raw data\18_01_Juveniles_arrested_Education.csv")
economic_df = pd.read_csv(r"C:\Users\Sai\OneDrive\Desktop\Raw data\18_02_Juveniles_arrested_Economic_setup.csv", delimiter="\t")

education_trend = education_df.groupby("Year")[
    ["Education_Illiterate", "Education_Upto_primary", 
     "Education_Above_Primary_but_below_Matric_or_Higher_Secondary", 
     "Education_Matric_or_Higher_Secondary_&_above"]
].sum()


plt.figure(figsize=(12, 6))
for column in education_trend.columns:
    plt.plot(education_trend.index, education_trend[column], marker="o", linestyle="-", label=column)

plt.xlabel("Year")
plt.ylabel("Number of Juveniles Arrested")
plt.title("Crime Rate Trends Across Different Education Levels Over the Years")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()


economic_trend = economic_df.groupby("Year")[
    ["Economic_Set_up_Annual_Income_upto_Rs_25000", "Economic_Set_up_Annual_Income_250001_to_50000", 
     "Economic_Set_up_Middle_income_from_50001_to_100000", "Economic_Set_up_Middle_income_from_100001_to_200000", 
     "Economic_Set_up_Upper_middle_income_from_200001_to_300000", "Economic_Set_up_Upper_income_above_Rs_300000"]
].sum()


plt.figure(figsize=(12, 6))
for column in economic_trend.columns:
    plt.plot(economic_trend.index, economic_trend[column], marker="o", linestyle="-", label=column)

plt.xlabel("Year")
plt.ylabel("Number of Juveniles Arrested")
plt.title("Crime Rate Trends Across Different Income Levels Over the Years")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```
